```python
import networkx as nx
import matplotlib.pyplot as plt
import uuid
```

```python
g = nx.DiGraph(id=uuid.uuid4())
a = uuid.uuid4()
b = uuid.uuid4()

g.add_node(
    a,
    type="tensor",
)
g.add_node(
    b,
    type="blockop"
)

g.add_edge(a, b, type="input")

print(g)

nx.drawing.draw_networkx(
    g,
    with_labels=True,
    labels={
        a: "<<table><tr><td>Test</td></tr><tr><td>Xyz</td></tr></table>>",
        b: "Bar",
    },
)

pd = nx.nx_pydot.to_pydot(g)

pd.write_png("example.png")
```
